```python
ix = 3
iu = 2
iw = 2
iq = 2
ip = 2
N = 30
tf = 3
delT = tf/N
```

```python
# time-invariant matrices
C = np.array([[0,0,1],[0,0,0]])
D = np.array([[0,0],[1,0]])
E = np.array([[1,0],[0,1],[0,0]])
G = np.zeros((iq,iw))
```

## obstacle setting

```python
def get_H_obs(rx,ry) :
    return np.diag([1/rx,1/ry])
# obstacle
c_list = []
H_list = []
c1 = [1,2]
H1 = get_H_obs(0.75,1.5)
c_list.append(c1)
H_list.append(H1)
c2 = [4,3]
H2 = get_H_obs(0.75,1.5)
c_list.append(c2)
H_list.append(H2)
```

## initial and final conditions 

```python
xi = np.zeros(3)
xi[0] = 0.0
xi[1] = 0.0 
xi[2] = 0

xf = np.zeros(3)
xf[0] = 5.0
xf[1] = 5.0
xf[2] = 0

xi_margin = np.array([0.1,0.1,np.deg2rad(5)])*2
Qf = np.diag([0.5**2,0.5**2,np.deg2rad(20)**2])*1.5
vec_neighbor = []
xi_neighbor = np.array(get_neighbor_vec(-1,0,ix,None,vec_neighbor)) * np.array(xi_margin) + xi

myModel = UnicycleModel.unicycle('Hello',ix,iu,iw,'numeric_central')
myCost = UnicycleCost.unicycle('Hello',ix,iu,N)
myConst = UnicycleConstraints.UnicycleConstraints('Hello',ix,iu)
myConst.set_obstacle(c_list,H_list)

x0 = np.zeros((N+1,ix))
for i in range(N+1) :
    x0[i] = (N-i)/N * xi + i/N * xf
u0 = np.zeros((N+1,iu))

A,B,s,z,x_prop_n = myModel.diff_discrete_zoh(x0[0:N,:],u0[0:N,:],delT,tf) 
S = np.eye(ix)
R = 2*np.eye(iu)
Kbar = get_K_discrete(A,B,S,R,S,N,ix,iu)
Qbar = np.tile(np.diag([0.35**2,0.35**2,np.deg2rad(10)**2]),(N+1,1,1))
Ybar = Kbar@Qbar[:N]
xbar,ubar = x0,u0
```

```python
plt.figure(0,figsize=(7,7))
plot_traj_set(xbar,ubar,c_list,H_list,Qbar,xi=xi,xf=xf,xi_margin=xi_margin,Qf=Qf,plt=plt)
```

```python
history = []
total_iter = 30
for idx_iter in range(total_iter) :
    # step 1 Nominal trajectory update
    i1 = Single_PTR('unicycle',N,tf,1,myModel,myCost,myConst,
              w_c=1,w_vc=1e3,w_tr=1e1,tol_vc=1e-6,tol_tr=1e-4,verbosity=True)
    _,_,xnew,unew,total_num_iter,flag_boundary,l,l_vc,l_tr = i1.run(xbar,ubar,xi,xf,Qbar,Kbar)
    # plt.figure(0,figsize=(7,7))
    # plot_traj_set(xnew,unew,c_list,H_list,Qbar,xi=xi,xf=xf,xi_margin=xi_margin,Qf=Qf,plt=plt)
    # plt.show()
 
    A,B,F,s,z,x_prop_n = myModel.diff_discrete_zoh_noise(xnew,unew,np.zeros((N,iw)),delT,tf) 
    sz = tf*s + z
    # propagation
    xprop,_ = i1.forward_multiple(xnew,unew)

    myM = Lipschitz(ix,iu,iq,ip,iw,N)
    myM.initialize(xnew,unew,xprop,Qbar,Kbar,A,B,C,D,E,F,G,myModel)
    gamma = myM.update_lipschitz_norm(myModel,delT)
#     gamma = np.load('gammatest.npy')
    print("mean of gamma",np.mean(gamma,0),"max of gamma",np.max(gamma,0),"var of gamma",np.var(gamma,0))

    # propagation error
    e_prop = np.linalg.norm(xprop - xnew[1:],axis=1)

    myQ = Q_update(ix,iu,iq,ip,iw,N,delT,myCost.S,myCost.R,w_tr=1e1,w_vc=1e4)
    myQ.initialize(xnew,unew,e_prop*0,A,B,C,D,E,F,G)
    alpha = 0.99
    Qnew,Knew,Ynew,status = myQ.solve_w_propagation_error(alpha,gamma,xi_neighbor,Qf,Qbar,Ybar)
    print("LMI status:" + status)

    # plot the trajectory
    plt.figure(1,figsize=(7,7))
    plot_traj_set(xnew,unew,c_list,H_list,Qnew,xi=xi,xf=xf,xi_margin=xi_margin,Qf=Qf,plt=plt)
    plt.show()

    # measure the difference
    xdiff = np.linalg.norm(xbar-xnew)
    udiff = np.linalg.norm(ubar-unew)
    Qdiff = np.linalg.norm(Qbar-Qnew)
    Kdiff = np.linalg.norm(Kbar-Knew)
    Ydiff = np.linalg.norm(Ybar-Ynew)

#     print("xdiff:{:8.5f} udiff:{:8.5f} Qdiff:{:8.5f} Kdiff:{:8.5f} Ydiff:{:8.5f}".format(xdiff,udiff,Qdiff,Kdiff,Ydiff))

    # update trajectory
    xbar = xnew
    ubar = unew
    Qbar = Qnew
    Kbar = Knew
    Ybar = Ynew
    
    # save trajectory
    traj = {}
    traj['x'] = xbar
    traj['u'] = ubar
    traj['Q'] = Qbar
    traj['Y'] = Ybar
    traj['K'] = Kbar
    traj['gamma'] = gamma
    traj['Q_diff'] = Qdiff
    traj['K_diff'] = Kdiff
    history.append(traj)
    
```

```python
Q_diff_list = [history[i]['Q_diff'] for i in range(idx_iter)]
K_diff_list = [history[i]['K_diff'] for i in range(idx_iter)]
```

```python
i_index = np.array([i+1 for i in range(idx_iter)])
fS = 20
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.plot(i_index,Q_diff_list,'o-')
# plt.plot(i_index,i_index*0+tol_Q,'--')
plt.xlabel('iteration number',fontsize=fS,fontname='Times New Roman')
plt.ylabel(r'$\Delta_Q$',fontsize=fS,fontname='Times New Roman')
plt.yscale('log')
# plt.xticks(i_index)
# plt.legend(fontsize=fS)
plt.subplot(122)
plt.plot(i_index,K_diff_list,'o-')
# plt.plot(i_index,i_index*0+tol_K,'--',label="tolerance")
plt.xlabel('iteration number',fontsize=fS,fontname='Times New Roman')
plt.ylabel(r'$\Delta_K$',fontsize=fS,fontname='Times New Roman')
plt.yscale('log')
# plt.xticks(i_index)
plt.rcParams["font.family"] = "Times New Roman"
plt.legend(fontsize=fS)

# plt.savefig('/Users/keemtw/Documents/Study/Note/SCP_invariant_tube/convergence_result.pdf')
```
